```python
import requests, bs4, re, time
```

```python
pagedata = requests.get("https://www.aerzteblatt.de/archiv")
```

```python
cleanpagedata = bs4.BeautifulSoup(pagedata.text,"html.parser")
```

```python
print(cleanpagedata)
```

```python
heftidsearch = re.compile(r'heftid=(\d+)')
heftids = heftidsearch.findall(str(cleanpagedata))
```

```python
print(heftids)
```

```python
heftids_set = set(heftids)
print(heftids_set)
```
